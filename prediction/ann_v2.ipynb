# Evaluation

* Class 0 = Slight severity
* Class 1 = Serious severity

## ROC AUC Scores
* AUC Score (One-vs-Rest): 0.701: This score represents the average AUC of each class against all other classes. An AUC of 0.701 suggests that the model has moderate discriminatory ability overall.
* AUC Score (One-vs-One): 0.701: This score represents the average AUC for each pair of classes. The same score of 0.701 here suggests that the model performs consistently when distinguishing between any two classes.

## Classification Report
### Precision:
Precision is the ratio of true positive predictions to the total predicted positives. High precision indicates that the model does not have many false positives.
* Class 0: Precision is 0.86, meaning that 86% of the instances predicted as Class 0 are correct.
* Class 1: Precision is 0.29, indicating a high number of false positives.
* Class 2: Precision is 1.00, but this is misleading because it is based on zero actual predictions (see confusion matrix).

### Recall:
Recall is the ratio of true positive predictions to the total actual positives. High recall indicates that the model captures most of the positive instances.

* Class 0: Recall is 1.00, meaning the model correctly identifies all instances of Class 0.
* Class 1: Recall is 0.00, indicating that the model fails to identify Class 1 instances.
* Class 2: Recall is 0.00, meaning the model fails to identify Class 2 instances.

### F1-Score: 
The F1-score is the harmonic mean of precision and recall. It balances the two metrics.

* Class 0: F1-score is 0.92, indicating strong performance.
* Class 1: F1-score is 0.00, indicating poor performance.
* Class 2: F1-score is 0.00, indicating poor performance.

### Support:
Support is the number of actual occurrences of each class in the test set.

* Class 0: 325002 instances.
* Class 1: 48379 instances.
* Class 2: 5110 instances.

## Confusion Matrix
The confusion matrix provides a more detailed look at how the model's predictions are distributed across the classes.

* Class 0 (Majority Class): Almost all instances (324977 out of 325002) are correctly classified, with only 25 misclassified as Class 1.
* Class 1: The model almost entirely fails to identify Class 1 instances, with 48369 misclassified as Class 0 and only 10 correctly classified.
* Class 2: The model completely fails to identify Class 2 instances, with all 5110 misclassified as Class 0.]]


## Summary
### Strengths:
The model is very good at identifying the majority class (Class 0), as indicated by the high precision, recall, and F1-score for this class.
### Weaknesses:
The model performs very poorly on the minority classes (Class 1 and Class 2), with almost no correct predictions for these classes. This indicates a significant class imbalance issue where the model is biased towards the majority class.

## Things we should have done different:

### Address Class Imbalance:
We could consider techniques like oversampling the minority classes, undersampling the majority class, or using class weights to balance the training process.
### Model Tuning:
Experiment with different architectures, hyperparameters, or alternative models that might handle class imbalance better.
